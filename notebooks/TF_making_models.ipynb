## Import data

```python
conn = sqlite3.connect('../datasets/database.db')
c = conn.cursor()
```

```python
query = """
        SELECT *
        FROM quake_table
        LEFT JOIN moons_table on moons_table.time = quake_table.time
    """
conn.row_factory = sqlite3.Row
c.execute(query)
quakes = c.fetchall()
```

```python
strong_tides = []
for row in quakes:
    if row[6] != None:
        strong_tides.append(row)
```

```python
print(len(quakes))
print(len(strong_tides))
```

```python
len(strong_tides)/len(quakes) * 100
```
